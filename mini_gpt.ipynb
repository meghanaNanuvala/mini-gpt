## Data Collection and Preprocessing

Used the `wikimedia/wikipedia` dataset from Hugging Face (20231101.en snapshot). Due to resource constraints, I selected the first 100,000 entries to build a manageable and diverse corpus. Each entry was cleaned and split into paragraph-level lines, ensuring removal of:

- Headings and markup tags (==Heading==, [[links]], etc.)
- Templates, tables, and other irrelevant sections
- Non-informative or repetitive content

```python
from datasets import load_dataset
import re

# Load Wikipedia
raw_dataset = load_dataset("wikimedia/wikipedia", "20231101.en", split="train")

# data of 100k
data_100k = raw_dataset.select(range(100000))

# Preprocessing function
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'\[\[.*?\]\]', '', text)
    text = re.sub(r'==.*?==+', '', text)
    text = re.sub(r'\{\{.*?\}\}', '', text)
    text = re.sub(r'\(.*?edit.*?\)', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'File:[\w\-.]+\.(jpg|png|gif)', '', text)
    return text.strip()

# Clean all entries
cleaned_texts = [clean_text(entry['text']) for entry in data_100k]


```

```python
# Save to .txt file
with open("cleaned_wiki_100k.txt", "w", encoding="utf-8") as f:
    for paragraph in cleaned_texts:
        if paragraph:
            f.write(paragraph.replace("\n", " ").strip() + "\n")
```

## Tokenizer Training

Trained a Byte-Pair Encoding (BPE) tokenizer using the Hugging Face `tokenizers` library. The tokenizer was trained from scratch on our preprocessed Wikipedia data with a vocabulary size of 20,000.

```python
!pip install -q tokenizers

from tokenizers import Tokenizer, models, trainers, pre_tokenizers, decoders, processors

# Initialize tokenizer
tokenizer = Tokenizer(models.BPE())

# Pre-tokenizer: splits on whitespace and punctuation
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# Special tokens
special_tokens = ["<pad>", "<unk>", "<s>", "</s>"]

# Trainer
trainer = trainers.BpeTrainer(
    vocab_size=20000,
    special_tokens=special_tokens
)

# Train from file
tokenizer.train(files=["cleaned_wiki_100k.txt"], trainer=trainer)

# Save tokenizer
tokenizer.save("bpe_tokenizer_100k.json")

tokenizer = Tokenizer.from_file("bpe_tokenizer_100k.json")

sample = "Anarchism is a political philosophy."
encoded = tokenizer.encode(sample)

print("Input Text:", sample)
print("Token IDs:", encoded.ids)
print("Tokens:", encoded.tokens)
print("Vocabulary Size:", tokenizer.get_vocab_size())
```

## Dataset Preparation

Each line in the dataset was encoded using the trained tokenizer, prepended with the `<s>` token, and split into fixed-length (128-token) chunks. We right-padded shorter sequences with `<pad>` and created input-target pairs for causal language modeling (CLM).

**Input Shape:** (861,981, 128)  
**Label Shape:** (861,981, 128)

```python
import torch
from torch.utils.data import Dataset, DataLoader

# Load tokenizer
from tokenizers import Tokenizer
tokenizer = Tokenizer.from_file("bpe_tokenizer_100k.json")

SEQ_LEN = 128
PAD_TOKEN_ID = tokenizer.token_to_id("<pad>")
BOS_TOKEN_ID = tokenizer.token_to_id("<s>")

# Load raw text lines
with open("cleaned_wiki_100k.txt", "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

# Encode each paragraph with BOS token
encoded_lines = []
for line in lines:
    tokens = tokenizer.encode("<s> " + line).ids  # add BOS token
    encoded_lines.extend(tokens + [PAD_TOKEN_ID])  # pad between sequences

# Split into chunks of SEQ_LEN
def chunk_sequence(data, seq_len):
    return [data[i:i+seq_len] for i in range(0, len(data), seq_len) if len(data[i:i+seq_len]) == seq_len]

chunks = chunk_sequence(encoded_lines, SEQ_LEN)

# Build input and target sequences
inputs = torch.tensor(chunks)
labels = inputs.clone()  # labels are just shifted inputs for CLM

# Shift labels to the right by one
labels = torch.roll(labels, shifts=-1, dims=1)

print("Input Example (token IDs):", inputs[0])
print("Label Example (token IDs):", labels[0])

# Save tensors 
torch.save(inputs, "inputs_100k.pt")
torch.save(labels, "labels_100k.pt")

print("Saved inputs.pt and labels.pt successfully.")
```

## Model Architecture and Training

We implemented a unidirectional Transformer decoder-only architecture (GPT-style) using Keras and keras-nlp.

```python
# !pip install -q keras-nlp tensorflow

import keras
import keras_nlp
import tensorflow as tf
from keras.callbacks import ModelCheckpoint
import torch
import numpy as np

# Load saved PyTorch tensors
inputs = torch.load("inputs_100k.pt")
labels = torch.load("labels_100k.pt")
inputs = inputs.numpy()
labels = labels.numpy()

print("inputs and labels loaded and converted to NumPy.")
print("Shape:", inputs.shape, labels.shape)

VOCAB_SIZE = 20000
SEQ_LEN = 128
EMBED_DIM = 256
NUM_HEADS = 4
NUM_LAYERS = 4
FEED_FORWARD_DIM = 1024
BATCH_SIZE = 32
EPOCHS = 2
PAD_TOKEN_ID = 0

# Strategy for Multi-GPU Support
strategy = tf.distribute.MirroredStrategy()
print(f"Number of GPUs: {strategy.num_replicas_in_sync}")
GLOBAL_BATCH_SIZE = BATCH_SIZE * strategy.num_replicas_in_sync

# Dataset Preparation
train_dataset = tf.data.Dataset.from_tensor_slices((inputs, labels))
train_dataset = train_dataset.shuffle(1000).batch(GLOBAL_BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Model Definition and Training
with strategy.scope():
    inputs_ = keras.layers.Input(shape=(SEQ_LEN,), dtype="int32")
    x = keras.layers.Masking(mask_value=PAD_TOKEN_ID)(inputs_)

    x = keras_nlp.layers.TokenAndPositionEmbedding(
        vocabulary_size=VOCAB_SIZE,
        sequence_length=SEQ_LEN,
        embedding_dim=EMBED_DIM,
        mask_zero=True
    )(x)

    for _ in range(NUM_LAYERS):
        x = keras_nlp.layers.TransformerDecoder(
            num_heads=NUM_HEADS,
            intermediate_dim=FEED_FORWARD_DIM,
        )(x)

    outputs = keras.layers.Dense(VOCAB_SIZE)(x)
    model = keras.Model(inputs=inputs_, outputs=outputs)

    loss_fn = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    perplexity = keras_nlp.metrics.Perplexity(from_logits=True, mask_token_id=PAD_TOKEN_ID)
    optimizer = keras.optimizers.Adam(learning_rate=0.0001, clipnorm=1.0)

    model.compile(optimizer=optimizer, loss=loss_fn, metrics=[perplexity])

    model.summary()

    # Train the model
    history = model.fit(train_dataset, epochs=EPOCHS)

    # Save the model
    model.save("gpt_mini_model_100k.keras")
```

## Evaluation and Text Generation

We evaluated the model on validation loss and perplexity, obtaining a perplexity of **100.76**, which is reasonable for a 20k-token vocabulary trained on 100k articles.


```python
import tensorflow as tf
from tokenizers import Tokenizer

# Load trained model and tokenizer
model = tf.keras.models.load_model("gpt_mini_model_100k.keras", compile=False)
tokenizer = Tokenizer.from_file("bpe_tokenizer_100k.json")

SEQ_LEN = 128
PAD_TOKEN_ID = tokenizer.token_to_id("<pad>")

# Greedy Generation Function
def generate_text(prompt, max_len=64):
    encoded = tokenizer.encode("<s> " + prompt).ids
    generated = encoded.copy()

    for _ in range(max_len - len(encoded)):
        # Get last SEQ_LEN tokens (truncate if needed)
        input_ids = generated[-SEQ_LEN:]
        padding_len = SEQ_LEN - len(input_ids)
        input_padded = input_ids + [PAD_TOKEN_ID] * padding_len

        input_tensor = tf.constant([input_padded], dtype=tf.int32)

        # Predict next token
        preds = model(input_tensor, training=False)
        next_token_logits = preds[:, len(input_ids) - 1, :]
        next_token_id = tf.argmax(next_token_logits, axis=-1).numpy()[0]

        # Stop if </s> token is predicted
        if next_token_id == tokenizer.token_to_id("</s>"):
            break

        generated.append(next_token_id)

    return tokenizer.decode(generated)
```

```python
prompts = [
    "the role of education",
    "climate change and its impact",
    "artificial intelligence is",
    "<s>"
]


print("📄 Generated Text Examples:")
for i, prompt in enumerate(prompts, 1):
    output = generate_text(prompt, max_len=64)
    print(f"\n🔹 Example {i} – Prompt: {prompt}")
    print(output)
```
